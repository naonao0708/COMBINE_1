```python
# 去除重复列函数
def remove_duplicate_columns(df):
    """
    移除DataFrame中的重复列名，只保留第一次出现的列。

    参数:
    df: pd.DataFrame, 原始DataFrame。

    返回:
    df_unique_columns: pd.DataFrame, 移除重复列名后的DataFrame。
    """
    # 查找重复的列名
    duplicated_columns = df.columns.duplicated()

    # 选择不重复的列
    df_unique_columns = df.loc[:, ~duplicated_columns]
    
    return df_unique_columns
```

```python
import pandas as pd
from sklearn.impute import KNNImputer

import pandas as pd
from scipy.stats import zscore
from sklearn.preprocessing import MinMaxScaler

import pandas as pd
from scipy.stats import zscore

def discretize_expression(data):
    # 计算Z-score
    z_scores = data.apply(zscore)
    
    # 离散化
    discretized_data = z_scores.applymap(lambda x: 1 if x > 1 else (-1 if x < -1 else 0))
    
    return discretized_data


def fill_missing_values_with_knn(df, n_neighbors=5):
    """
    使用K近邻算法填充DataFrame中的缺失值。

    参数:
    df: pd.DataFrame, 原始带有缺失值的DataFrame。
    n_neighbors: int, 用于KNNImputer的邻居数，默认为3。

    返回:
    df_filled: pd.DataFrame, 缺失值被填充后的DataFrame，保留原始的行列标签。
    """
    # 确保传入的是DataFrame
    if not isinstance(df, pd.DataFrame):
        raise ValueError("输入必须是一个pandas DataFrame。")
    
    # 创建KNNImputer实例
    imputer = KNNImputer(n_neighbors=n_neighbors)

    # 进行拟合和转换，填充缺失值
    imputed_data = imputer.fit_transform(df)

    # 将填充后的数据转换回DataFrame格式，并保留原始的行和列标签
    df_filled = pd.DataFrame(imputed_data, columns=df.columns, index=df.index)

    return df_filled
```
